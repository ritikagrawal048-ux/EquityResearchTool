```python
#Gemini Version

import os
import pickle
import time
import streamlit as st

from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import UnstructuredURLLoader
from langchain_community.vectorstores import FAISS

# from langchain.chains import RetrievalQAWithSourcesChain

from langchain_google_genai import (
    ChatGoogleGenerativeAI,
    GoogleGenerativeAIEmbeddings,
)
```

```python
#load openAI api key
os.environ['GEMINI_API_KEY'] = 'YOUR_API_KEY_HERE'
```

```python
# Initialise LLM with required params
#llm = OpenAI(temperature=0.9, max_tokens=500)   #OPEN AI

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.5
)
```

```python
!pip3 install unstructured libmagic python-magic python-magic
```

# (1) Load data

```python
loaders = UnstructuredURLLoader(urls=[
    "https://www.moneycontrol.com/news/business/markets/wall-street-rises-as-tesla-soars-on-ai-optimism-11351111.html",
    "https://www.moneycontrol.com/news/business/tata-motors-launches-punch-icng-price-starts-at-rs-7-1-lakh-11098751.html"
])
data = loaders.load() 
len(data)
```

# (2) Split data to create chunks

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

# As data is of type documents we can directly use split_documents over split_text in order to get the chunks.
docs = text_splitter.split_documents(data)
```

```python
len(docs)
```

```python
docs[0]
```

## (3) Create embeddings for these chunks and save them to FAISS index


```python
!pip install faiss-cpu
```

```python
#Create the embeddings
embeddings = GoogleGenerativeAIEmbeddings(
    model="gemini-embedding-001"
)

#Creating the FAISS Index for vector search
vector_index = FAISS.from_documents(
    docs,
    embeddings
)
```

```python
vector_index
```

```python
#Save
vector_index.save_local("/kaggle/working/faiss_index_1")
```

```python
# Load
vector_index = FAISS.load_local(
    "/kaggle/working/faiss_index_1",
    embeddings,
    allow_dangerous_deserialization=True
)
```

```python
vector_index
```

## (4) Retrieve similar embeddings for a given question and call LLM to retrieve final answer

```python
pip install langchain-classic
```

```python
#Step1 : Create prompt template
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template("""
Answer the question using only the provided context.

Context:
{context}

Question:
{input}
""")

#Step2: Create document chain
from langchain_classic.chains.combine_documents import create_stuff_documents_chain

document_chain = create_stuff_documents_chain(
    llm,
    prompt
)


#Step3: Create the retriever
retriever = vector_index.as_retriever()


#Step4: Create the retrieval chain
from langchain_classic.chains import create_retrieval_chain

retrieval_chain = create_retrieval_chain(
    retriever,
    document_chain
)

```

```python
#Step5: Create the question
input_query = "what is the price of Tiago iCNG?"
response = retrieval_chain.invoke({
    "input": input_query
})

print(response["answer"])
```
